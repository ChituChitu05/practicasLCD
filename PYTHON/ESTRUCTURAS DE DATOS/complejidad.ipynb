# Orden de complejidad O( ).

## Qué es
    El orden de complejidad indica cómo crece el tiempo o la memoria de un algoritmo cuando aumenta el tamaño de la entrada 𝑛.
    No mide segundos reales, mide crecimiento.
## Órdenes de complejidad más comunes.
* n es la cantidad de datos.
### O(1) — constante:
    No depende de 𝑛.
    Ejemplo: acceder a una posición de un arreglo.
    

    def obtener_primero(lista):
        return lista[0]
### O(log n) — logarítmica:
    El problema se reduce a la mitad cada paso.
    Ejemplo: búsqueda binaria.
    Muy eficiente para datos grandes.
    
    def busqueda_binaria(lista, x):
    inicio = 0
    fin = len(lista) - 1


    while inicio <= fin:
        medio = (inicio + fin) // 2
        if lista[medio] == x:
            return True
        elif lista[medio] < x:
            inicio = medio + 1
        else:
            fin = medio - 1
    return False

### O(n) — lineal:
    Crece proporcional a 𝑛.
    Ejemplo: recorrer una lista completa.


    def buscar(lista, x):
    for elemento in lista:
        if elemento == x:
            return True
    return False

### O(n log n) — casi lineal:
    Común en algoritmos eficientes de ordenamiento.
    Ejemplo: Merge Sort, Quick Sort (promedio).


    def merge_sort(lista):
        if len(lista) <= 1:
            return lista

        medio = len(lista) // 2
        izquierda = merge_sort(lista[:medio])
        derecha = merge_sort(lista[medio:])

        return merge(izquierda, derecha)

    def merge(izq, der):
        resultado = []
        i = j = 0

        while i < len(izq) and j < len(der):
            if izq[i] < der[j]:
                resultado.append(izq[i])
                i += 1
            else:
                resultado.append(der[j])
                j += 1

        resultado.extend(izq[i:])
        resultado.extend(der[j:])
        return resultado

### O(n²) — cuadrática:
    Dos ciclos anidados.
    Ejemplo: comparar todos con todos.

    def pares(lista):
        for i in lista:
            for j in lista:
                print(i, j)

### Mala para datos grandes:
    (2ⁿ) / O(n!) — exponencial.
    Crecimiento explosivo.
    Ejemplo: fuerza bruta, permutaciones.
    Solo sirve para 𝑛 pequeño.
